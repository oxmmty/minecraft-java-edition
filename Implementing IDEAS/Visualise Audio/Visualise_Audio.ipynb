```python
import matplotlib.pylab as plt

def save_plot(filename):
    y, sr = librosa.load(filename)        
    plt.figure()
    plt.plot(y)
    plt.xlabel('time')
    plt.ylabel('amplitude')
    plt.plot()
    plt.savefig('audio.png')

save_plot('falling.mp3')
```
